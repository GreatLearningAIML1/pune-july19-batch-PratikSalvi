### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
k=0
while(k<len(cols)-1):
    max=Iris[cols[k]].max()
    min=Iris[cols[k]].min()
    range=max-min
    print('Range of ',cols[k], ' is ',range)
    k+=1
    
    print('done')
```

### Find out the no of male & female employees in the company

*Hint: You can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
#Considered 1 as male and 2 as female
male=0
female=0
for i in HR['Gender']:
    if i==1:
        male=male+1
    else:
        female=female+1
            
print('Males are',male)
print('Females are',female)
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
q3,q1=np.percentile(HR['Age'],[75,25])
print(q3)
print(q1)
IQR=q3-q1
print('IQR is -',IQR)
```
